```python
import pandas as pd
from scipy.stats import spearmanr
import numpy as np

autoencoder_results_df = pd.read_excel('F:/Uni Y3/tesi/arbeaut.xlsx')

final_product_df = pd.read_excel('F:/Uni Y3/tesi/fp4.xlsx')
```

```python
autoencoder_results_df
```

```python
# extract the patient IDs from the first column of autoencoder_results_df
autoencoder_patient_ids = autoencoder_results_df.iloc[:, 0]

# extract the patient IDs from the first row of final_product_df
gene_expression_patient_ids = final_product_df.iloc[0, 0:].index

# filter out non-unique patient IDs
autoencoder_patient_ids_unique = autoencoder_patient_ids.unique()
autoencoder_patient_ids_unique = [id.split('.')[0] for id in autoencoder_patient_ids_unique]
gene_expression_patient_ids_unique = gene_expression_patient_ids.unique()

# find common patient IDs between autoencoder results and gene expression
common_patient_ids = set(autoencoder_patient_ids_unique).intersection(gene_expression_patient_ids_unique)
common_patient_ids_list = list(common_patient_ids)

# extract numerical part of each ID and convert to integers
numerical_ids = [int(patient_id.split('-')[2]) for patient_id in common_patient_ids_list]

# sort patient IDs based on numerical parts
common_patient_ids = sorted(common_patient_ids, key=lambda x: (int(x.split('-')[1]), int(x.split('-')[2])))

print(common_patient_ids)
```

```python
correlation_results = []

for row_index in range(len(final_product_df)):
    # get the gene expression values
    gene_row = final_product_df[common_patient_ids]
    gene_row_values = gene_row.iloc[row_index, 0:].values 
    for col_index in range(1, autoencoder_results_df.shape[1]):  # Skip the first column
        # get the autoencoder feature values
        autoencoder_values = autoencoder_results_df.iloc[:, col_index].values
        #calculate the Spearman correlation between the autoencoder feature and gene expression values
        correlation, p_value = spearmanr(autoencoder_values, gene_row_values)
        # store  results if correlation is greater than 0.6
        if correlation > 0.6:
            correlation_results.append((autoencoder_results_df.columns[col_index], correlation, p_value, row_index))
  
```

```python
for feature, corr, p_value, row_index in correlation_results:
    print(f"Feature {feature} has correlation of {corr} with the gene {row_index}")
```

```python
# save df to excell
import os
df = pd.DataFrame(correlation_results, columns=['Feature', 'Correlation', 'P-Value', 'Row_Index'])

directory = "F:/Uni Y3/tesi"
file_path = f"{directory}/corrres.xlsx"

if not os.path.exists(directory):
    os.makedirs(directory)

df.to_excel(file_path, index=False)
```
