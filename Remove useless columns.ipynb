```python
import pandas as pd

file_path = r'F:\Uni Y3\tesi\botfeatimg\img\botfeatres.xlsx'
df = pd.read_excel(file_path)

# if more than half the colums have a 0 value drop it
columns_to_drop = [col for col in df.columns if (df[col] == 0).sum() > len(df) / 2]
df = df.drop(columns=columns_to_drop)

# save file
output_file_path = r'F:\Uni Y3\tesi\botfeatimg\botfeatres2.xlsx'
df.to_excel(output_file_path, index=False)

print(f"Columns dropped: {columns_to_drop}")
print(f"Modified DataFrame saved to {output_file_path}")
```
