```python
import pandas as pd
import numpy as np

file_path = 'F:/Uni Y3/tesi/botfeatimg/img/botfeatres.xlsx'
data_df = pd.read_excel(file_path)

# skip first column and row
data_df = data_df.iloc[1:, 1:]

# convert the data to numeric, drop nan
data_df = data_df.apply(pd.to_numeric, errors='coerce').dropna()

# calculate general statistics for the entire dataset
general_mean = data_df.values.mean()
general_std_dev = data_df.values.std()

# identify the column with the highest standard deviation then sort and print top 10
column_with_highest_std = data_df.std().idxmax()
highest_std_value = data_df.std().max()
sorted_std_devs = data_df.std().sort_values(ascending=False)
top_10_columns_with_highest_std = sorted_std_devs.head(10)

print(f"General mean of the dataset: {general_mean}")
print(f"General standard deviation of the dataset: {general_std_dev}")
print(f"Column with the highest standard deviation: '{column_with_highest_std}' with a standard deviation of {highest_std_value}")

print("\nTop 10 columns with the highest standard deviation:")
print(top_10_columns_with_highest_std)
```
